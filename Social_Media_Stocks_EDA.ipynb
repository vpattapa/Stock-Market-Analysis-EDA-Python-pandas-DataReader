```python
fig, axs = plt.subplots(nrows=3, ncols=1, figsize=(8, 8), sharex=False)

q1_m = SM_data['Close', 'META'].quantile(0.25)
q3_m = SM_data['Close', 'META'].quantile(0.75)
iqr_m = q3_m - q1_m
filter_m = (SM_data['Close', 'META'] >= q1_m - 1.5 * iqr_m) & (SM_data['Close', 'META'] <= q3_m + 1.5 * iqr_m)
filtered_data_m = SM_data.loc[filter_m, ('Close', 'META')]
axs[0].plot(filtered_data_m)
axs[0].set_title('Filtered Meta Closing Price')

q1_pins = SM_data['Close', 'PINS'].quantile(0.25)
q3_pins = SM_data['Close', 'PINS'].quantile(0.75)
iqr_pins = q3_pins - q1_pins
filter_pins = (SM_data['Close', 'PINS'] >= q1_pins - 1.5 * iqr_pins) & (SM_data['Close', 'PINS'] <= q3_pins + 1.5 * iqr_pins)
filtered_data_pins = SM_data.loc[filter_pins, ('Close', 'PINS')]
axs[1].plot(filtered_data_pins)
axs[1].set_title('Filtered Pinterest Closing Price')

q1_snap = SM_data['Close', 'SNAP'].quantile(0.25)
q3_snap = SM_data['Close', 'SNAP'].quantile(0.75)
iqr_snap = q3_snap - q1_snap
filter_snap = (SM_data['Close', 'SNAP'] >= q1_snap - 1.5 * iqr_snap) & (SM_data['Close', 'SNAP'] <= q3_snap + 1.5 * iqr_snap)
filtered_data_snap = SM_data.loc[filter_snap, ('Close', 'SNAP')]
axs[2].plot(filtered_data_snap)
axs[2].set_title('Filtered Snapcaht Closing Price')


# Rotate the x-axis labels for each subplot
for ax in axs.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.tight_layout()
plt.show()
```

**Calculating stock return : Return = (close-open)/open**

```python
SM_data['Return', 'META'] = (SM_data['Close', 'META']  - SM_data['Open', 'META'])/SM_data['Open', 'META']
SM_data['Return', 'PINS'] = (SM_data['Close', 'PINS']  - SM_data['Open', 'PINS'])/SM_data['Open', 'PINS']
SM_data['Return', 'SNAP'] = (SM_data['Close', 'SNAP']  - SM_data['Open', 'SNAP'])/SM_data['Open', 'SNAP']
```

### Analysis on the returns of the stock
The return provides a clear depiction of a stock's growth over time. It reflects the potential profit (or loss) one might realize if they were to purchase the stock at the opening of the trading day and sell it at the close. The return plot offers a visual representation of these fluctuations, highlighting days that yield a profit and those that result in a loss, thereby illustrating the stock's daily performance and investment volatility.

### Insights from Stock Returns Histograms

- **Volatility Analysis**: The spread of the histograms indicates the volatility of each stock. A wider spread suggests higher volatility. 
- **Return Frequency**: The height of the histogram bars represents the frequency of returns within a particular range. 
- **Skewness and Risk**: The skewness of the histograms (left or right) provides insight into the risk profile of the stock. 
- **Comparative Return Tendencies**: Positioning relative to the zero line indicates whether a stock typically has positive or negative returns. 
- **Diversification Strategy**: By comparing the return distributions, investors can strategize on diversifying their portfolio. Combining stocks with varying levels of volatility can help in achieving a balanced investment approach.

These histograms serve as a visual tool for assessing historical performance and can be a part of the decision-making process for potential future investments. It's crucial to complement these observations with a thorough fundamental analysis, as historical trends do not guarantee future outcomes.

**But usually we will analyze a short period of returns like 3 months or 6 months to analyze how the company is performing**

```python
#Plotting histogram of the returns
bins = 10
fig, axs = plt.subplots(3, 1, figsize=(10, 10), sharex=False)
axs[0].hist(SM_data['Return', 'META'].dropna(), bins=bins, color='blue', edgecolor='black', alpha=0.7)
axs[0].set_title('META Returns')

axs[1].hist(SM_data['Return', 'PINS'].dropna(), bins=bins, color='Red', edgecolor='black', alpha=0.7)
axs[1].set_title('Pinterest Returns')

axs[2].hist(SM_data['Return', 'SNAP'].dropna(), bins=bins, color='Green', edgecolor='black', alpha=0.7)
axs[2].set_title('Snapchat Returns')

plt.tight_layout()
plt.show()
```

### META: 
The histogram displays a slight left skew, indicating a modest inclination towards negative returns. However, the concentration of data around the center suggests moderate volatility.

### Pinterest: 
This histogram has a right skew with a broader spread, implying more days with positive returns but also a higher volatility compared to META.

### Snapchat: 
The returns show a clear right skew with a wide distribution, signaling higher volatility. There's a noticeable chance for substantial positive returns, pointing towards greater risk and reward potential.

In summary, META appears to exhibit moderate stability, Pinterest shows growth with variability, and Snapchat offers high-risk-high-reward returns.

```python
#Differencing
SM_data['Diff', 'META'] = SM_data['Close', 'META'].diff(1)
SM_data['Diff', 'PINS'] = SM_data['Close', 'PINS'].diff(1)
SM_data['Diff', 'SNAP'] = SM_data['Close', 'SNAP'].diff(1)
```

```python
#Plotting Difference

fig, axs = plt.subplots(3, 1, figsize=(10, 10), sharex=False)
axs[0].plot(SM_data['Diff', 'META'])
axs[0].set_title('Meta Difference in closing Price')
axs[1].plot(SM_data['Diff', 'PINS'])
axs[1].set_title('Pinterest Difference in closing Price')
axs[2].plot(SM_data['Diff', 'SNAP'])
axs[2].set_title('Snapchat Difference in closing Price')

for ax in axs.flat:
    for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.tight_layout()
plt.show()
```
